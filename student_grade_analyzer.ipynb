```python
student_name=input("Enter Student name:")
grade=[]
for i in range(3):
    mark = int(input(f"Enter grade for subject {i+1}: "))
    grade.append(mark)
student_detail={'name':student_name,
                   'grade':grade}
def calculate_average(grade):
    total=sum(grade)
    average=total/len(grade)
    return average
def assign_grade(average):
    if average>=90:
        return "A"
    elif average>=70:
        return "B"
    elif average>=50:
        return "C"
    else:
        return "D"
def check_result(average):
    if average>=40:
        return "Pass"
    else:
        return "Fail"
average = calculate_average(grade)
grade = assign_grade(average)
result = check_result(average)

print("\nStudent Name:", student_detail['name'])
print("Average Score:", average)
print("Grade:", grade)
print("Result:", result)
```
